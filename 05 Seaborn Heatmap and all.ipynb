```python
#Now by looking at Heatmap we can clearly see that some columns are unwanted
#so to remove such columns, first select them
columns = game_wrangled.columns.tolist()
print(columns)
```

```python
#now to delete the columns
#just like wine.drop['quality'], but here more than 1 column so
columns = [c for c in columns if c not in['bayes_average_rating','average_rating','type','name','id','minplayers','maxplayers','yearpublished','playingtime','minplaytime','maxplaytime','minage']]

#but we need target as average_rating so saving it
#we droped average_rating in column because we need to predict it (think)
target = 'average_rating'

columns
```

```python
#generate training and test datasets
from sklearn.model_selection import train_test_split

#generate training set
train = game_wrangled.sample(frac=0.8,random_state=1)
test = game_wrangled.loc[game_wrangled.index.isin(train.index)]

print(train.shape)
print(test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

LR = LinearRegression()
LR.fit(train[columns],train[target])

predicted = LR.predict(test[columns])

print(mean_squared_error(test[target],predicted))
```
