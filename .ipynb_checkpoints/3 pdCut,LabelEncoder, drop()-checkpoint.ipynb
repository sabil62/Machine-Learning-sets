```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wine_dataset = pd.read_csv('winequality-red.csv',sep=';')
```

```python
wine_dataset.head(8)
```

```python
#wine_database.describe()
#wine_database.shape
```

```python
Bins =[2,5,8]
group_name = ['bad','good']
wine_dataset['quality'] = pd.cut(wine_dataset['quality'],bins = Bins, labels = group_name)
```

```python
wine_dataset.head(10)
```

```python
#doing groupby just to test
#wine_groupByQuality = wine_dataset.groupby('quality')
#wine_groupByQuality.sum()
    #wine_groupByQuality.mean()
```

```python
from sklearn.preprocessing import LabelEncoder
label_quality = LabelEncoder()
```

```python
wine_dataset['quality'] = label_quality.fit_transform(wine_dataset['quality'])
```

```python
wine_dataset.head()           #see now the quality is 0 or 1, instead of good or bad
```

```python
wine_dataset['quality'].value_counts()
```

# ML STARTS FROM HERE

```python
x = wine_dataset.drop('quality',axis=1)
y = wine_dataset['quality']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
#SVC
from sklearn import svm
svm_model = svm.SVC()
svm_model.fit(x_train,y_train)
```

```python
y_predict = svm_model.predict(x_test)

#y_predict
#y_test
```

```python
#to check accuracy
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print(classification_report(y_test,y_predict))
```
