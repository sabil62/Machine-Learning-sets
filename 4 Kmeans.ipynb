```python
import numpy as np
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as plt
```

```python
np.random.randint(50)
```

```python
mean1 = [np.random.randint(50),np.random.randint(50)]
mean2 = [np.random.randint(50),np.random.randint(50)]

#making diagonal covariance ??
cov = [[50,0],[0,50]]

x1,y1 = np.random.multivariate_normal(mean1,cov,50).T
x2,y2 = np.random.multivariate_normal(mean2,cov,50).T
```

```python
x1
```

```python
x = np.append(x1,x2)
y = np.append(y1,y2)

plt.plot(x,y,':')   #: le . . . wala feel hunxa
plt.axis('equal')
plt.show()
```

```python
X = np.array(zip(x,y))
```

# USING KMEANS

```python
import pandas as pd
kmeans = KMeans(n_clusters=2)

Data = {'x':x,
        'y':y}
df = pd.DataFrame(Data,columns=['x','y'])
```

```python
kmeans.fit(df)
```

```python
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
labels = kmeans.labels_
labels
```

```python
colour_choice=['y.','r.']

plt.scatter(df['x'], df['y'],c=labels.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
```
